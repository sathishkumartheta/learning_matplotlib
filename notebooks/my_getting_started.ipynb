<h2  style=background:#ffff00;border-style:solid;color:#003300>  <center>Imports</center></h2>

```python
import matplotlib.pyplot as plt
import numpy as np

from matplotlib.ticker import MultipleLocator, AutoMinorLocator
```

<h2 style=border-style:solid;color:#003300;background:#FFFF00> <center>Figure Size</center></h2>

```python
np.random.seed(33)
X=np.linspace(0.5,3.5,100)
Y1=3+np.cos(X)
Y2=1+np.cos(1+X/0.75)/2
Y3=np.random.uniform(Y1,Y2,len(X))

figure1=plt.figure(figsize=(7.5,7.5))

ax1=figure1.add_axes([0.2,0.17,0.68,0.7])


ax1.set_title('Anatomy of a Figure', fontsize='20', verticalalignment='bottom')
ax1.set_xlabel('x Axis label',fontsize='15')
ax1.set_ylabel('y Axis label',fontsize='15')


ax1.set_xlim([0,4])
ax1.set_ylim([0,4])

ax1.xaxis.set_major_locator(MultipleLocator(1.000))
ax1.xaxis.set_minor_locator(AutoMinorLocator(4))

ax1.yaxis.set_major_locator(MultipleLocator(1.0))
ax1.yaxis.set_minor_locator(AutoMinorLocator(4))

ax1.xaxis.set_minor_formatter("{x:0.2f}")
ax1.yaxis.set_minor_formatter("{x:0.2f}")

ax1.tick_params(which='major', width=1.0, length=10, labelsize=14, labelcolor='red')
ax1.tick_params(which='minor', width=1.0, length=5, labelsize=10, labelcolor='blue' )

ax1.grid(which='major',linestyle='--',color='.25',zorder=-10)

ax1.plot(X,Y1, label='blue signal')
ax1.plot(X,Y2, label='orange signal')
ax1.plot(X[::3],Y3[::3], marker='s', linewidth=0, markerfacecolor='none', markeredgecolor='C4', markeredgewidth=2.5)

ax1.legend(loc="upper right", fontsize='14')

```
